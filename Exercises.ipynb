```python
# Day 1 - Exercise 1

def Factorial(num):
  if num == 1:
    return 1
  else:
    return factorial(num-1) * num
arg = int(input("Enter a number to find its factorial: "))
print(Factorial(arg))
```

```python
# Day 1 - Exercise 2

class Agent():

  def get_response(self, text):
    pass

  # Returns boolean whether agent can handle input
  def can_handle(self, text):
    pass

class FunTechBot(Agent):
  def __init__(self):
    self.agents = []

  def add_agent(self, agent):
    self.agents.append(agent)

  def get_response(self, text):
    for agent in self.agents:
      if agent.can_handle(text):
        return agent.get_response(text)
      else:
        return "Sorry, I don't understand."

  class NumericAgent(Agent):

    def get_response(self, text):
      number = int(text)
      return str(number + 20)

    def can_handle(self, text):
      return text.isdigit()

FunTechBot = FunTechBot()
FunTechBot.add_agent(FunTechBot.NumericAgent())

user_input = input("You: ")
while user_input != "":
  print(f"FunTechBot: {FunTechBot.get_response(user_input)}")
  user_input = input("You: ")


```

```python
# Day 2 - Exercise 2
import io
from google.cloud import vision


client = vision.ImageAnnotatorClient().from_service_account_file("creds.json")

with io.open("landmark.jpg", "rb") as image_file:
  content = image_file.read()

image = vision.Image(content=content)
response = client.landmark_detection(image = image)
labels = response.landmark_annotations

for label in labels:
  print(label.description)
```

```python
with io.open("people.jpg", "rb") as image_file:
  content = image_file.read()

image = vision.Image(content=content)
response = client.face_detection(image=image)
faces = response.face_annotations

# Names of likelihood from google.cloud.vision.enums
likelihood_name = ('UNKNOWN', 'VERY_UNLIKELY', 'UNLIKELY', 'POSSIBLE',
                   'LIKELY', 'VERY_LIKELY')
for face in faces:
  print('anger: {}'.format(likelihood_name[face.anger_likelihood]))
  print('joy: {}'.format(likelihood_name[face.joy_likelihood]))
  print('surprise: {}\n'.format(likelihood_name[face.surprise_likelihood]))
```

```python
from google.cloud import language_v1

client = language_v1.LanguageServiceClient().from_service_account_file("creds.json")

text = input("Enter sentence: ")

document = language_v1.Document(content=text, type=language_v1.Document.Type.PLAIN_TEXT)

response = client.analyze_sentiment(request={'document': document, 'encoding_type': language_v1.EncodingType.UTF8})
sentences = response.sentences

for sentence in sentences:
  print(f"Sentence: {sentence.text.content}")
  print(f"Sentiment: {sentence.sentiment.score}\n")
```
