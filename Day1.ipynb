```python
import io
from google.cloud import vision

client = vision.ImageAnnotatorClient.from_service_account_json("creds.json")

with io.open("cat.png", 'rb')as image_file:
  content = image_file.read()

image = vision.Image(content=content)

response = client.label_detection(image=image)
labels = response.label_annotations

for label in labels:
  print(label.description)
```

```python
from google.cloud import language
from google.cloud.language_v1 import types

client = language.LanguageServiceClient.from_service_account_json("creds.json")
text = "I love cake"
document = types.Document(content=text, type=types.Document.Type.PLAIN_TEXT)

response = client.analyze_sentiment(document=document, encoding_type="UTF32")
sentences = response.sentences

for sentence in sentences:
  print(f"Sentence: {sentence.text.content}")
  print(f"Sentiment score: {sentence.sentiment.score}")
  print(f"Sentiment magnitude: {sentence.sentiment.magnitude}")
```

```python
from PIL import Image, ImageOps  # Install pillow instead of PIL
import numpy as np

# Disable scientific notation for clarity
np.set_printoptions(suppress=True)

# Try installing tf-keras for compatibility if model was saved with TensorFlow 2.x Keras
from tf_keras.models import load_model

# Load the model
model = load_model("keras_model.h5", compile=False)

# Load the labels
class_names = open("labels.txt", "r").readlines()

# Create the array of the right shape to feed into the keras model
# The 'length' or number of images you can put into the array is
# determined by the first position in the shape tuple, in this case 1
data = np.ndarray(shape=(1, 224, 224, 3), dtype=np.float32)

# Replace this with the path to your image
image = Image.open("cat.png").convert("RGB")

# resizing the image to be at least 224x224 and then cropping from the center
size = (224, 224)
image = ImageOps.fit(image, size, Image.Resampling.LANCZOS)

# turn the image into a numpy array
image_array = np.asarray(image)

# Normalize the image
normalized_image_array = (image_array.astype(np.float32) / 127.5) - 1

# Load the image into the array
data[0] = normalized_image_array

# Predicts the model
prediction = model.predict(data)
index = np.argmax(prediction)
class_name = class_names[index]
confidence_score = prediction[0][index]

# Print prediction and confidence score
print("Class:", class_name[2:], end="")
print("Confidence Score:", confidence_score)
```

```python
from keras.models import Sequential
from keras.layers import Dense, Activation, Dropout #

model = Sequential()
model.add(Dense(64,input_shape=(4,))) # Sets the amount of neurons,
                                      # and the amount of data entered

model.add(Activation('relu')) # Type of activation function that changes how the
                              # model trains/learns

model.add(Dropout(0.2)) # Adds regularisation. This is adding variation to
                         # prevent overfitting

model.add(Dense(64))

model.add(Activation('relu'))

model.add(Dense(2)) # Output node, here has two catagories (dog/cat)

model.add(Activation('softmax')) # Classification algo w/o linear relation
                                 # between classes
model.summary()
```

```python
import csv
import numpy as np
from tensorflow.keras.utils import to_categorical
from sklearn.model_selection import train_test_split
```

```python
from sklearn.model_selection import train_test_split

# Array that will be fed in
label = []
feature = []

# open csv file and create reader
with open("banknotes.csv") as csv_file:
  csv_reader = csv.reader(csv_file)
  line = 0
  for row in csv_reader:
    if line != 0:  ## Ignore headers
      feat=[]
      feat.append(float(row[0]))
      feat.append(float(row[1]))
      feat.append(float(row[2]))
      feat.append(float(row[3]))
      feature.append(feat)
      label.append(row[4]) # Last row contains classifications

    line += 1

feature = np.array(feature)
label = np.array(label)
label = to_categorical(label)

x_train, x_test, y_train, y_test = train_test_split(feature, label, test_size=0.15)
```

```python
model = Sequential()
model.add(Dense(64,input_shape=(4,)))
model.add(Activation('relu'))
model.add(Dropout(0.2))
model.add(Dense(64))
model.add(Dense(2, activation='softmax'))

# Loss checks correctness, optimizer uses loss value to train accuracy (adamn is algo)
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(x_train, y_train, epochs=23)  # epochs is the amount it runs
```

```python
result = model.evaluate(x_test, y_test)
print("Test Loss: ", result[0])
print("Test Accuracy: ", result[1])
```
